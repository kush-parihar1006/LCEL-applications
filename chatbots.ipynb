```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store = {}

def get_session_history(session_id: str) -> BaseChatMessageHistory:   ### this fucntion will check and retrieve chat history for a session id
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

with_message_history = RunnableWithMessageHistory(model, get_session_history)  ### this will create a wrapper around the model to manage chat history based on session id
```

```python
config = {"configurable":{"session_id":"chat1"}}  
## this is a configuration of chat history with that particular session and a conversation.
```

```python
with_message_history.invoke(
    [HumanMessage(content = "Hi I am Kush, I am a Generative AI engineer.")],
    config=config
    )
```

```python
response = with_message_history.invoke(
    [HumanMessage(content = "Whats my name and profession?")],
    config=config
    )

response.content
```

### Managing your history is really important since we don't want to overflow the LLM with loads of information and messages. Hence we need to set a limit to the message size that are parsed to the model.


##### trim_messages() - trims the input given to the llm model without losing essential data for the model to understand the requirements and revert back with the best possible answer.

```python
#%pip install transformers
```

```python
from langchain_core.messages import SystemMessage, trim_messages
trimmer = trim_messages(
    max_tokens = 850,
    strategy = "last",
    token_counter = model,
    include_system = True,
    allow_partial = False,
    start_on = "human"
)

messages = [
    SystemMessage(content = "You are a gym trainer and dietician with 10 years of experience"),
    HumanMessage(content = "Hi, I am Kush, I am 23 yrs old."),
    AIMessage(content = "Nice to meet you, Kush. What can I help you with today?"),
    HumanMessage(content = "Give me a diet plan for weight loss?"),
    AIMessage(content = "Sure, I can help you with that. I need to know a bit more about you and your goals."),
    HumanMessage(content = "I am 6 feet tall, I weigh 96 kg and I am a non-vegetarian but I don't eat red meat."),
    AIMessage(content = "Thanks for this info. What is your weight loss goal and fitness goal?"),
    HumanMessage(content = "I want to become 78 kgs with a good amount of muscle mass. Goal is to have a shredded physique with good stamina."),
    AIMessage(content = "Great and What is your current activity level?"),
    HumanMessage(content = "I am very lethargic and I don't exercise currently."),
    AIMessage(content = "Thanks for sharing all this info. Based on what you've told me, I would recommend the following diet plan for weight loss and muscle gain: \n\n1. Protein: Aim for at least 1.6-2.2 grams of protein per kilogram of body weight per day. This means you should be consuming around 154-211 grams of protein daily. Good sources of protein include chicken breast, turkey, fish, eggs, dairy products, and plant-based options like tofu and legumes.\n\n2. Carbohydrates: Focus on complex carbohydrates such as whole grains, fruits, and vegetables. These will provide you with sustained energy throughout the day. Aim for around 2-3 grams of carbohydrates per kilogram of body weight per day, which would be around 192-288 grams daily.\n\n3. Fats: Include healthy fats in your diet from sources like avocados, nuts, seeds, and olive oil. Aim for around 0.5-1 gram of fat per kilogram of body weight per day, which would be around 48-96 grams daily.\n\n4. Meal Timing: Consider spreading your meals throughout the day to keep your metabolism active and to provide a steady supply of nutrients to your muscles. Aim for 4-6 smaller meals rather than 2-3 large meals.\n\n5. Hydration: Drink plenty of water throughout the day to stay hydrated and support your metabolism.\n\n6. Supplements: You may also consider supplements like whey protein or creatine to support your muscle gain goals, but it's best to consult with a healthcare professional before starting any supplementation.\n\nRemember that consistency is key when it comes to achieving your fitness goals. Along with this diet plan, incorporating regular exercise such as strength training and cardio will help you reach your desired physique and improve your stamina.")
]
trimmer.invoke(messages) 
```

```python
## To Use this trimmer in the chain
from operator import itemgetter 
from langchain_core.runnables import RunnablePassthrough

chain = (
    RunnablePassthrough.assign(messages=itemgetter("messages") | trimmer)
    | prompt
    | model
)

response = chain.invoke({"messages": messages + [HumanMessage(content = "What is my name, age,height and weight?")],
                         "height":"6 feet",
                         "weight":"96 kg",
                         "food_type":"non-vegetarian but no red meat."})

response.content
```

```python
## Wrap it with message history to maintain the context of conversation and also trim the messages when the token limit exceeds.
with_message_history = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="messages"
)

config = {"configurable":{"session_id":"chat5"}}
```

```python
response = with_message_history.invoke(
    {"messages":[HumanMessage(content = "What is my name, age")],
     "height":"6 feet",
     "weight":"96 kg",
     "food_type":"non-vegetarian but no red meat."},
    config = config
)
response.content
```

### using retrievers, chains and vector databases
- The information is stored in the databases where it is retrieved using a retriever when similar questions are the asked to the LLM model.
- It can also be used to store documents containing varied information that will be used to train the model and answer similar questions precisely.

```python
#%pip install langchain-chroma
```

#### Langchain documents should be used to ensure that the model reads and answers data precisely when similar questions are asked. There are two types of components of a document.
- page_content : defines the string representing content of docs.
-metadata : information about the document. Information like connection with other documents, source of document and other useful information. 


```python
from langchain_core.documents import Document

documents = [
    Document(
        page_content="Fitness is essential for maintaining a healthy lifestyle and improving overall well-being.",
        metadata={"source": "fitness-doc"},
    ),
    Document(
        page_content="To remain fit and in shape, it is essential to exercise and move your body on a daily basis for atleast an hour .",
        metadata={"source": "exercise-doc"},
    ),
    Document(
        page_content="Additionally, maintaining a balanced diet is also crucial for achieving fitness goals and supporting overall health.",
        metadata={"source": "nutrition-doc"},
    ),
    Document(
        page_content="While Dieting, it is important to hit your macros and achieve your caloric goals for the desired results.",
        metadata={"source": "Macros-doc"},
    ),
    Document(
        page_content="Remember to get enough sleep and manage stress levels, as both can significantly impact your fitness journey and overall health.",
        metadata={"source": "Rest-doc"},
    ),
]
```

```python
documents 
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
from langchain_groq import ChatGroq
groq_api_key=os.getenv("GROQ_API_KEY")

os.environ["HF_TOKEN"]=os.getenv("HF_TOKEN")

llm=ChatGroq(groq_api_key=groq_api_key,model="llama-3.1-8b-instant")
llm
```

The data must be stored in the form of vectors. Hence the texts is converted into vectors before storing the data inside vector databases. This embedding technique of text into vectors is carried using open source hugging-face model.

```python
%pip install sentence-transformers
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
### vectore databases - chromadb
from langchain_chroma import Chroma
vectorstore=Chroma .from_documents(documents,embedding=embeddings)
vectorstore
```

```python
vectorstore.similarity_search('fitness')
```

```python
## async method
await vectorstore.asimilarity_search('fitness')
```

```python
vectorstore.similarity_search_with_score('fitness')
```

### Retrievers that will be used to extract the relevant information from the chromaDB

```python
from typing import List 

from langchain_core.documents import Document
from langchain_core.runnables import RunnableLambda 

retriever = RunnableLambda(vectorstore.similarity_search).bind(k=2)
retriever.batch(["fitness","exercise"])
```

### Method 2 - Vectore store implement as a Retriever. VectoreStoreRetriever. We create a retriever that will use different attributes and paramters to get a method for best extraction of data. This is majorly used at an industry level

```python
retriever = vectorstore.as_retriever(
    search_type = "similarity",
    search_kwargs = {"k":2}
)
retriever.batch(["fitness","exercise"])
```

#### Integrate the retriever with the chain 

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough

message = """
Answer this question using the provided context only.

{question}

Context:
{context}
"""
prompt = ChatPromptTemplate.from_messages([("human", message)])

rag_chain={"context":retriever,"question":RunnablePassthrough()}|prompt|llm

response=rag_chain.invoke("How does exercise and dieting help me in weight loss?")
print(response.content)
```

Maintaining a balanced diet and hitting your macros, while also exercising, is crucial for achieving fitness goals and supporting overall health.
